## Reading the Global Temperature and ONI data 

I downloaded global datasets for ONI and Global Temperatures 

Specifically these are

- ONI: [NOAA PSL Ocean Nino Index](https://climatedataguide.ucar.edu/climate-data/nino-sst-indices-nino-12-3-34-4-oni-and-tni). This data contain computed monthly ONI anomalies
- [Berkeley Global Temperature Anomaly (1850 to Recent)](https://berkeleyearth.org/data/)

The code below reads these data into a pandas dataframes and reformats them to be a simple timeseries. 

```python
ONI_data = '../Data/oni.data'
oni_data = pd.read_csv(ONI_data, sep=r'\s+', skiprows=1, header = None, skipfooter=8, index_col=0) 
oni_data.head()
```

```python
df_oni= oni_data.stack() # converts the dataframe into a series with a multi-index of Year and Month 
df_oni= pd.DataFrame(df_oni) # converts the series back into a dataframe
df_oni.index.rename(['Year','Month'],inplace=True) # renames the multi-index levels to Year and Month
df_oni=df_oni.reset_index() # resets the index so that Year and Month become columns instead of index levels
df_oni.columns=['Year','Month','ONI'] # renames the columns to Year, Month, and ONI
df_oni['Date'] = pd.to_datetime(df_oni['Year'].astype(str) + '-' + df_oni['Month'].astype(str), format='%Y-%m') # converts year and month into a datetime object   
df_oni = df_oni.set_index('Date').drop(columns=['Year','Month'])# sets the date as the index and drops the Year and Month columns
df_oni.head()
```

```python
berkeley_temperature_data = '../Data/Global_TAVG_monthly.txt'

df_temp = pd.read_csv(berkeley_temperature_data, sep=r'\s+', skiprows=95, header=None, 
                      names=['Year', 'Month', 'T_month_anomaly', 'T_month_uncertainty'], 
                      usecols=[0, 1, 2, 3])
df_temp=df_temp.reset_index(drop=True) # resets the index to be a simple range index
df_temp['Date'] = pd.to_datetime(df_temp['Year'].astype(str) + '-' + df_temp['Month'].astype(str), format='%Y-%m')
df_temp = df_temp.set_index('Date').drop(columns=['Year','Month'])# sets the date as the index and drops the Year and Month columns
df_temp.head()
```
